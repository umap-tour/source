```python
from glob import glob
import json
from natsort import natsorted
import sys
import re

pattern = re.compile('epoch(\d+)-batch(\d+)')
dir_in = './googlenetTrain-umap-51200D-to-15D-50000-maxpool 2/'


epoch_arch_to_dir = {}

epoch0 = 0
batch = 0
for fn in natsorted(glob(f'{dir_in}/epoch*-batch*')):
    if epoch != epoch0:
        batch = 0
        epoch0 = epoch
    epoch, _ = pattern.findall(fn)[0]
    print(epoch, batch)
    epoch_arch_to_dir[f'{epoch},{batch}'] = fn
    batch += 1
    
print(json.dumps(epoch_arch_to_dir, indent=2))
```
